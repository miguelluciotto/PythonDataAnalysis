## Primeros Gráficos

Creamos un gráfico de línea.

```python
x = [1, 2, 3, 4, 5]
y = [2, 6, 9, 4, 11]

plt.plot(x, y)

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Line Graph')

plt.show()
```
